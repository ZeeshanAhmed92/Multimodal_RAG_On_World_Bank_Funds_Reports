```python
import pymupdf, pymupdf4llm

doc = pymupdf.open("./source_docs/2020TrustFundAnnualReports.pdf")
hdr = pymupdf4llm.IdentifyHeaders(doc, max_levels=3)
md = pymupdf4llm.to_markdown(doc, hdr_info=hdr)
```

```python
print(md)
```

```python
with open("output.md", "w", encoding="utf-8") as f:
    f.write(md)
```
